```python
# Define a function to calculate totals
def calculate_totals(df):
    total_entries = len(df.iloc[:, 3])  # Total entries in the first column
    non_nan_count = df.iloc[:, 3].notna().sum()  # Count of non-NaN values in the first column
    nan_count = total_entries - non_nan_count # Count of NaN values in the first column
    return total_entries, non_nan_count, nan_count

# Calculate totals for each file
totals_2APIC5 = calculate_totals(Apic2_5)

#print(totals_APIC3, totals_APIC4, totals_APIC5)
print(totals_2APIC5)
```

```python
labels = ['2 APIC 5']
total_entries = totals_2APIC5[0]
non_nan_counts = totals_2APIC5[1]
nan_counts = totals_2APIC5[2]

# Set up the bar plot
x = range(len(labels))  # X-axis positions
width = 0.2  # Width of the bars
bar_spacing = 0.05  # Space between the bar groups

plt.figure(figsize=(10, 6))
plt.bar(x, total_entries, width, label='Nombre des étudiants', color='#54473F', align='center')
plt.bar([p + width for p in x], non_nan_counts, width, label='Nombre des présences', color='#AB886D', align='center')
plt.bar([p + 2*width for p in x], nan_counts, width, label='Nombre des absences', color='#AFB86D', align='center')

# Remove the top and right borders
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

# Add labels and title
plt.xlabel('Class')
plt.ylabel('Nombre des étudiants')
plt.title('Nombre total des étudiants et nombre total des présences par class')
plt.xticks([p + width / 2 for p in x], labels)  # Center the ticks between the bars
plt.legend()

# Show the plot
plt.tight_layout()
plt.show()
```
